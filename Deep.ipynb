```python
tensor_dependent = tf.constant(y)
```

```python
import keras
from keras import layers
```

```python
%%time
model = keras.Sequential([
    keras.layers.Dense(3),
    keras.layers.Dense(10,activation=keras.activations.relu),
    keras.layers.Dense(5,activation=keras.activations.relu),
    keras.layers.Dense(10,activation=keras.activations.relu),
    keras.layers.Dense(1,activation=None)
    ])

model.compile(loss=keras.losses.mean_absolute_error,
              optimizer=keras.optimizers.Adam(),
              metrics=['mae'])
model.fit(tensor,tensor_dependent,epochs=10,batch_size=32,validation_split=0.2)
```

```python
tensor[0],tensor_dependent[0]
```

```python
tensor_res = np.array([[ 1.03162451, -0.82518364, -0.46544089], 
                   [ 0.57035972, -0.37399065, -2.51886534],
                   [ 1.22445585,  0.60717939, -1.22840413]])

# Predykcja na podstawie tych danych
predictions = model.predict(tensor_res)
```
